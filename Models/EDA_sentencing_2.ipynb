```python
def get_contrast_score(df, gender, statut, degree,target_col):
    s_1 = df[(df['statut']==statut)&\
                    (df['clfely']==degree)&\
                    (df['gender']==gender)&\
                   (df['race']=='BLACK')&\
                   (df['total']<= 1000)][target_col]

    s_2 = df[(df['statut']==statut)&\
                        (df['clfely']==degree)&\
                        (df['gender']==gender)&\
                       (df['race']=='WHITE')&\
                       (df['total']<= 1000)][target_col]
    
    fig, ax = plt.subplots(1, 1,figsize=(6,4))

    colors = ['#aacfcf', '#d291bc']

    sns.boxplot(x='series',y='values',
                data=pd.DataFrame({'values':pd.concat([s_1,s_2],axis=0),
                                  'series':np.repeat(["Black","White"],[len(s_1),len(s_2)])}),
                ax = ax,palette=colors,width=0.5
               )
```

```python
get_contrast_score(df = df_sentence_t, 
            gender = 'MALE',
            statut = '893.13(6)(A)',
            degree = '3RD DEGREE',
            target_col = "total")
```

```python
get_contrast_score(df = df_sentence_t, 
            gender = 'MALE',
            statut = '810.02(4)',
            degree = '3RD DEGREE',
            target_col = "total")
```

```python
get_contrast_score(df = df_sentence_t, 
            gender = 'MALE',
            statut = '812.014(2)(C)6',
            degree = '3RD DEGREE',
            target_col = "total")
```

# Sentencing length

#### Sentencing length variation within same crime, race comparison

```python
def get_contrast_length(df, gender, statut, degree,target_col):
    s_1 = df[(df['statut']==statut)&\
                    (df['clfely']==degree)&\
                    (df['gender']==gender)&\
                   (df['race']=='BLACK')][target_col]

    s_2 = df[(df['statut']==statut)&\
                        (df['clfely']==degree)&\
                        (df['gender']==gender)&\
                       (df['race']=='WHITE')][target_col]
    
    fig, ax = plt.subplots(1, 1,figsize=(6,4))

    colors = ['#aacfcf', '#d291bc']

    sns.boxplot(x='series',y='values',
                data=pd.DataFrame({'values':pd.concat([s_1,s_2],axis=0),
                                  'series':np.repeat(["Black","White"],[len(s_1),len(s_2)])}),
                ax = ax,palette=colors,width=0.5
               )
```

```python
get_contrast_length(df = df_sentence_t, 
            gender = 'MALE',
            statut = '943.0435(9)',
            degree = '3RD DEGREE',
            target_col = "sp_total")
```

```python
get_contrast_length(df = df_sentence_t, 
            gender = 'MALE',
            statut = '943.0435(9)',
            degree = '2ND DEGREE',
            target_col = "sp_total")
```

# Relation between sentencing length and total points

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df_sentence_reg = df_sentence_t.dropna()
```

```python
df_sentence_reg = df_sentence_reg[df_sentence_reg["clfely"]!= 'CLFELY']
```

```python
df_sentence_t.shape
```

```python
df_sentence_reg.shape
```

```python
mod = smf.ols(formula='sp_total ~ I(race) + I(gender) + I(clfely) + total', data=df_sentence_reg)

res = mod.fit(cov_type='HC3')

print(res.summary())
```

```python
df_b = df_sentence_reg[df_sentence_reg['race']=='BLACK']
df_w = df_sentence_reg[df_sentence_reg['race']=='WHITE']
```
